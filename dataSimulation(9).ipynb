```python
from scipy.stats import uniform
from numpy import ones,vstack
from numpy.linalg import lstsq

def getRandomPointsOnEdge(p1, p2, n=20):
    # delete n random points on edge by uniform distribution
    
    def y(x, m, c):
        return m * x + c
    
    points = [p1, p2]
    x_coords, y_coords = zip(*points)
    A = vstack([x_coords,ones(len(x_coords))]).T
    m, c = lstsq(A, y_coords)[0]
    # Line Solution is y = mx + c
    if isclose(abs(p1[0] - p2[0]), 0, abs_tol=0.00001):
        yMax, yMin = max(p1[1], p2[1]), min(p1[1], p2[1])
        L = yMax - yMin
        y_delete = yMin + L * np.random.rand(n)
        x_delete = [p1[0]] * n
    else:
        xMax, xMin = max(p1[0], p2[0]), min(p1[0], p2[0])
        L = xMax - xMin
        x_delete = xMin + L * np.random.rand(n)
        y_delete = np.array([y(x, m, c) for x in x_delete])
    
    return (x_delete, y_delete)

p1 = (hull_x[0], hull_y[0])
p2 = (hull_x[1], hull_y[1])

print(p1, p2)
x_delete, y_delete = getRandomPointsOnEdge(p1, p2)

fig, ax = plt.subplots(1, 2, figsize=(15, 4))

# ax[0].plot(noise_x, noise_y, 'ro', c='red', alpha=1, markersize=2)
ax[0].plot(hull_x, hull_y)
ax[0].plot(x_delete, y_delete, 'ro', c='white', alpha=1, markersize=2)
ax[0].grid(True)


ax[1].plot([p1[0], p2[0]], [p1[1], p2[1]], alpha=1, markersize=3)
ax[1].plot(x_delete, y_delete, 'ro', c='red', alpha=1, markersize=3)
# ax[1].plot(noisePoint[0], noisePoint[1], 'ro', c='blue', alpha=1, markersize=5)
ax[1].grid(True)

plt.show()
```

```python
# https://github.com/stepmat/rectilinear_polygon_generator/blob/master/generate_polygon.py

import sys
import cv2
import numpy
import copy
import scipy.misc
import itertools
from PIL import Image, ImageOps, ImageDraw
from scipy.ndimage import morphology, label
from copy import deepcopy
from operator import itemgetter
from statistics import median, mean
from math import sqrt
from random import randint

def scalePolygon(poly_x, poly_y, size=(512.0, 512.0), padding=(30.0, 30.0)):
    # all coordinates are not negative! 
    xMin = np.min(poly_x)
    yMin = np.min(poly_y)   
    poly_x -= xMin
    poly_y -= yMin
    
    xMax = np.max(poly_x)
    yMax = np.max(poly_y)
    
    poly_x /= xMax
    poly_y /= yMax
    poly_x *= size[0] - 2 * padding[0]
    poly_y *= size[1] g- 2 * padding[1]
    poly_x += padding[0]
    poly_y += padding[1]
    return poly_x, poly_y

def convertToIntegerCoordinates(x, y):
    x = [int(el) for el in x]
    y = [int(el) for el in y]
    return x, y

def generateRectilinearPolygon(num_rect = 3,
                               canvas_shapes = (20, 20),
                               min_size = 1,
                               max_size = 5,
                               increments = 100,
                               area_m2 = 1000):
#     num_rect = randint(5, 15) # rect. .nums
    canvas_width, canvas_height = canvas_shapes[1], canvas_shapes[0]
    img = np.zeros([increments*canvas_height,increments*canvas_width,3],dtype=np.uint8)
    img.fill(255)
    img = Image.fromarray(img)
    img = ImageOps.grayscale(img)
    im = np.array(img)
    
    im = morphology.grey_dilation(im, (3, 3)) - im      # inner morphological gradient
    img = Image.fromarray(im)
    visual = img.convert('RGB')
    draw = ImageDraw.Draw(visual)

    # Generate random number of rectangles (must be touching) of random sizes
    boxes = []
    for i in range(num_rect):
        valid = 0
        while(valid == 0):
            width = randint(min_size, max_size)*increments
            height = randint(min_size, max_size)*increments
            pos_x = randint(max_size+1, canvas_width-max_size-1)*increments
            pos_y = randint(max_size+1, canvas_height-max_size-1)*increments
            for j in boxes:
                touching = 1
                if (pos_x+(width/2.0) < j[0]-(j[2]/2.0)):
                    touching = 0
                if (pos_x-(width/2.0) > j[0]+(j[2]/2.0)):
                    touching = 0
                if (pos_y+(height/2.0) < j[1]-(j[3]/2.0)):
                    touching = 0
                if (pos_y-(height/2.0) > j[1]+(j[3]/2.0)):
                    touching = 0
                if touching == 1:
                    valid = 1
            if len(boxes) == 0:
                valid = 1
        boxes.append([pos_x,pos_y,width,height])
    # for i in boxes:
    #     print(i)
    # print('******************')


    # Extend those that stick out to the ground
    bottom_ground = 0
    for i in boxes:
        if (i[1] + (i[3]/2.0)) > bottom_ground:
            bottom_ground = (i[1] + (i[3]/2.0))

    for i in range(len(boxes)):
        bottom_left = [boxes[i][0]-(boxes[i][2]/2.0),boxes[i][1]+(boxes[i][3]/2.0)]
        bottom_right = [boxes[i][0]+(boxes[i][2]/2.0),boxes[i][1]+(boxes[i][3]/2.0)]
        line_check = [bottom_left,bottom_right]
        intersect = 0
        for j in range(len(boxes)):
            if i!=j:
                p1 = [boxes[j][0]-(boxes[j][2]/2.0),boxes[j][1]-(boxes[j][3]/2.0)-1]
                p2 = [boxes[j][0]-(boxes[j][2]/2.0),boxes[j][1]+(boxes[j][3]/2.0)+1]
                p3 = [boxes[j][0]+(boxes[j][2]/2.0),boxes[j][1]-(boxes[j][3]/2.0)-1]
                p4 = [boxes[j][0]+(boxes[j][2]/2.0),boxes[j][1]+(boxes[j][3]/2.0)+1]
                line_check2 = [p1,p2]
                line_check3 = [p3,p4]
                if (line_check[0][0] < line_check2[0][0]) and (line_check[1][0] > line_check2[0][0]):
                    if (line_check[0][1] > line_check2[0][1]) and (line_check[0][1] < line_check2[1][1]):
                        intersect = 1
                if (line_check[0][0] < line_check3[0][0]) and (line_check[1][0] > line_check3[0][0]):
                    if (line_check[0][1] > line_check3[0][1]) and (line_check[0][1] < line_check3[1][1]):
                        intersect = 1
                if (line_check[0][0] > line_check2[0][0]) and (line_check[1][0] > line_check2[0][0]):
                    if (line_check[0][0] < line_check3[0][0]) and (line_check[1][0] < line_check3[0][0]):
                        if (line_check[0][1] > line_check2[0][1]) and (line_check[0][1] < line_check2[1][1]):
                            intersect = 1
        if intersect == 0:
            current_bottom = boxes[i][1] + (boxes[i][3]/2.0)
            to_add = bottom_ground-current_bottom
            boxes[i][3] = boxes[i][3] + to_add
            boxes[i][1] = boxes[i][1] + (to_add/2.0)

    # Draw resulting rectangles as image
    tlbr = []
    for i in boxes:
        top_left = (i[0]-(i[2]/2.0),i[1]-(i[3]/2.0))
        bottom_right = (i[0]+(i[2]/2.0),i[1]+(i[3]/2.0))
        tlbr.append((top_left, bottom_right))
        draw.rectangle((top_left,bottom_right), fill='white')
    
    img = np.array(visual.convert('L'))
#     img = cv2.imread('output.jpg',0)
    ret,thresh = cv2.threshold(img,127,255,0)
    contours,hierarchy = cv2.findContours(thresh, 1, 2) 
    cnt = contours[0]
    M = cv2.moments(cnt)

    epsilon = 0.001 * cv2.arcLength(cnt,True)
    approx = cv2.approxPolyDP(cnt,epsilon,True)
    poly_x = approx[:, :, 0]
    poly_x = np.append(poly_x, approx[:, :, 0][0])
    poly_y = approx[:, :, 1]
    poly_y = np.append(poly_y, approx[:, :, 1][0])
    
    polygon = Polygon([(p[0],p[1]) for p in zip(poly_x, poly_y)])
    polygonArea = polygon.area # in pix^2
    polygonArea /= meter2pixels**2 # in metres
    
    proportion = area_m2 / polygonArea
    proportion_side = np.sqrt(proportion)
    
    poly_x = proportion_side * poly_x
    poly_y = proportion_side * poly_y

    return poly_x, poly_y

poly_x, poly_y = generateRectilinearPolygon(randint(5, 15), area_m2=1000)
poly_x, poly_y = scalePolygon(poly_x, poly_y)
plt.plot(poly_x, poly_y)
```

```python
NOISE_COVER_EDGE_PART = 0.5 # part of perimeter
NOISE_COVER_INNER_PART = 0.0008 # part of inner polygon area
INNER_POINT_DENSITY = (20, 80) #want: 5-10 px/m^2
EDGE_POINT_DENSITY = (60, 100) #want: 20-100 px/m^2
EDGE_POINT_STD = (2, 6)
INNER_POINT_STD = (2, 3)


def getCoordsOnLineFromStart(p1, p2, c):
    L = getLengthOfLine(p1, p2)
    x = ((L-c)*p1[0] + c*p2[0]) / L
    y = ((L-c)*p1[1] + c*p2[1]) / L
    return (x, y)

def getPerimeterPolygon(poly_x, poly_y):
    polygon = Polygon([(p[0],p[1]) for p in zip(poly_x, poly_y)])
    return polygon.length

def getAreaPolygon(poly_x, poly_y):
    polygon = Polygon([(p[0],p[1]) for p in zip(poly_x, poly_y)])
    polygonArea = polygon.area # in pix^2
    return polygonArea

def generateRandomPointsOnPolygonUniformly(poly_x, poly_y, numPoints):
    # generate Perimeter(poly) // NOISE_STEP points     
    polygon = [(p[0],p[1]) for p in zip(poly_x, poly_y)]
    edgesLen = [0.0]
    for i in range(len(poly_x)-1):
        p1, p2 = (poly_x[i], poly_y[i]), (poly_x[i+1], poly_y[i+1])
        l = getLengthOfLine(p1, p2)
        edgesLen.append(l)
    preSum = np.cumsum(edgesLen)
    perimeter = preSum[-1]
#     numPoints = int(perimeter * NOISE_COVER_PART)
    vals = np.random.uniform(0.0, perimeter, numPoints)
    vals = sorted(vals)
    
    x, y = [], []
    
    edgeInd = 1
    for delPoint in vals:
        while delPoint > preSum[edgeInd]:
            edgeInd += 1
        lenFromStart = delPoint - preSum[edgeInd - 1]
        coords = getCoordsOnLineFromStart(polygon[edgeInd-1], polygon[edgeInd], lenFromStart)
        x.append(coords[0])
        y.append(coords[1])

    return x, y


def generateNormalNoise(noise_centers_x, noise_centers_y, numPointsLim, stdLim):
    # there are we generate point  with density P1     
    x, y = [], []
    for i in range(len(noise_centers_x)):
        coords = (noise_centers_x[i], noise_centers_y[i])
        numPoints = np.random.randint(numPointsLim[0], numPointsLim[1] + 1)
        std = stdLim[0] + np.random.rand() * (stdLim[1] - stdLim[0])
        # std - also define linewidth, small values - thin, big values - thick
        # std = 5
        gaussNoise_x, gauusNoise_y = addNormalNoiseAroundPoint(coords, n=numPoints, sigma=std)
        x = x + list(gaussNoise_x)
        y = y + list(gauusNoise_y)
    return x, y

def generateRandomPointsInsidePolygon(poly_x, poly_y, numPoints):
    polygon = Polygon([(p[0],p[1]) for p in zip(poly_x, poly_y)])
#     n = int((polygonArea / (meter2pixels**2)) * INNER_POINT_DENSITY)
    xMin, xMax = np.min(polygon.exterior.xy[0]), np.max(polygon.exterior.xy[0])
    yMin, yMax = np.min(polygon.exterior.xy[1]), np.max(polygon.exterior.xy[1])
    
    innerNoise_x = []
    innerNoise_y = []
    for i in range(numPoints):
        inside = False
        point = None
        while not inside:
            random_x = np.random.uniform(xMin, xMax)
            random_y = np.random.uniform(yMin, yMax)
            point = Point(random_x, random_y)
            inside = polygon.contains(point)
        innerNoise_x.append(point.x)
        innerNoise_y.append(point.y)
    return innerNoise_x, innerNoise_y

# def generateNoiseInsidePolygon(centers_x, centers_y):
#     pass

def drawImageByPixels(x, y, imgSize=(512, 512)):
    arr = np.zeros(imgSize, dtype=np.uint8)
    for i in range(len(x)):
        if x[i] < imgSize[0] and y[i] < imgSize[1]:
            arr[x[i]][y[i]] = 255
        else:
            print('meow')
    img = Image.fromarray(arr)
    plt.imshow(img)
    plt.show()
    
# 1. generate polygon
poly_x, poly_y = generateRectilinearPolygon()

# 2. scale coordinates to 512x512
poly_x, poly_y = scalePolygon(poly_x, poly_y, size=(512.0, 512.0), padding=(50.0, 50.0))

# 3. generate noise centers (edge)
poly_perimeter = getPerimeterPolygon(poly_x, poly_y)
numNoiseCenters = int(poly_perimeter * NOISE_COVER_EDGE_PART)
edgeCent_x, edgeCent_y = generateRandomPointsOnPolygonUniformly(poly_x, poly_y, numNoiseCenters)

# 4. generate noise around noise centers (edge)
edgeNoise_x, edgeNoise_y = generateNormalNoise(edgeCent_x, edgeCent_y, EDGE_POINT_DENSITY, EDGE_POINT_STD)

# 5. generate random (uniform) noise centers (inner points of polygon)
poly_area = getAreaPolygon(poly_x, poly_y)
numInnerNoiseCenters = int(poly_area * NOISE_COVER_INNER_PART)
innerCent_x, innerCent_y = generateRandomPointsInsidePolygon(poly_x, poly_y, numInnerNoiseCenters)

# 6. generate noise around noise centers (inner)
innerNoise_x, innerNoise_y = generateNormalNoise(innerCent_x, innerCent_y, INNER_POINT_DENSITY, INNER_POINT_STD)

# 7. convert all coordinates to integer values
edgeNoise_x, edgeNoise_y = convertToIntegerCoordinates(edgeNoise_x, edgeNoise_y)
innerNoise_x, innerNoise_y = convertToIntegerCoordinates(innerNoise_x, innerNoise_y)

# 8. draw images by pixels
drawImageByPixels(edgeNoise_x + innerNoise_x, edgeNoise_y + innerNoise_y)
drawImageByPixels(innerNoise_x, innerNoise_y)
drawImageByPixels(edgeNoise_x, edgeNoise_y)

# checking EDGE_DENSITY_POINT
# num_exp = 100
# density_exp = 0
# for exp in range(num_exp):
#     x, y = generateRandomPointsOnPolygonUniformly(poly_x, poly_y, 1)
#     x, y = convertToIntegerCoordinates(x, y)
#     x = x[0]
#     y = y[0]

#     arr = np.zeros((512, 512), dtype=np.uint8)
#     for i in range(len(noise_x)):
#         if noise_x[i] < 512 and noise_y[i] < 512:
#             arr[noise_x[i]][noise_y[i]] = 255

#     arr1 = arr[x-5:x+5,y-5:y+5]
#     exp_res = (arr1.sum()/255)
#     density_exp += exp_res
# print(density_exp / num_exp)
# #     img = Image.fromarray(arr1)
#     plt.imshow(img)
#     plt.show()
```
